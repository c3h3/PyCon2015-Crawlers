```python
from selenium import webdriver
```

```python
browser = webdriver.Firefox()
```

```python
browser.get("http://railway.hinet.net/ctno1.htm")
```

```python
from IPython.display import Image
```

```python
Image(data=browser.get_screenshot_as_png())
```

```python
personId = browser.find_element_by_id("person_id")
```

```python
personId.send_keys("A176613941")
```

```python
from_station = browser.find_element_by_id("from_station")
```

```python
from selenium.webdriver.support.ui import Select
select_from_station=Select(from_station)
```

```python
select_from_station.select_by_visible_text(u"100-台北")
```

```python
select_to_station = Select( browser.find_element_by_id("to_station"))
select_to_station.select_by_visible_text(u"175-台南")
```

```python
browser.find_element_by_css_selector("button[type=\"submit\"]").click()
```
