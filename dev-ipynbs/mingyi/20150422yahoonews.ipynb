##來用正規表示法找出最後一頁

```python
import re #regular expression
```

```python
n=re.search('[1-9][0-9]*',lastpage)
m=n.group(0)
m
```

##<b style="color:##FF9999">來寫個一次抓吧</b>

```python
c=[]
for i in range(int(m)):
    url="https://tw.news.yahoo.com/politics/archive/%s.html"%(i+1)
    html = urlopen(url).read()#urllib
    S=PyQuery(html)
    Tit=S("div.story a span")
    len(Tit)
    for i in range(len(Tit)):
            c.append(Tit[i].text)
dfs=pd.DataFrame(c)
dfs    
```

```python
dfs['Date']="2015/04/22" #我知道自己偷塞是不對的
```
