```python
res=requests.get("http://www.amazon.com/s/ref=nb_sb_noss_2?url=search-alias%3Daps&field-keywords=data+analysis")
```

```python
from pyquery import PyQuery
```

```python
S=PyQuery(res.text)
s=S("li.s-result-item")
```

```python
ss=PyQuery(s[0])
```

```python
ss('h2').text()
```

```python
ss=[]
c=[]
for i in range(10):
    i+=1
    page=i
    url="http://www.amazon.com/s/ref=sr_pg_%s?page=%s&keywords=data+analysis"%(page,page)
    res=requests.get(url)
    S=PyQuery(res.text)
    s=S("li.s-result-item")
    ss=[PyQuery(i) for i in s]
    for y in ss:
        c.append(y('h2').text())
c        
        
    
```

```python
import pandas as pd
```

```python
dfs=pd.DataFrame(c)
```

```python
dfs.columns=['title']
```
