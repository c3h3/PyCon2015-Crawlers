##啟動pyquery

````python
from pyquery import PyQuery
````

````python
s=PyQuery(res.text)
````

###可以選到多個東西裡面的其中一個的方法

````python
S=s('.yt-lockup-title a')#S.attr只能找到一個
T0=PyQuery(S[0])#想要選到東西就得再包一層
T1=PyQuery(S[1])
print T0.attr('href')
print T1.attr('href')
````

````python
T0.attr('href')
````

````python
T1.attr('href')
````

````python
C1=URL+T0.attr('href')#頻道連結
C2=URL+T1.attr('href')
````

再來最後要搜尋連結了

````python
S1=PyQuery(requests.get(C1).text)
S2=PyQuery(requests.get(C2).text)
````

````python
print S1.text()
````

##整理第一筆，第二筆也一樣

````python
SS=S1('.pl-video-title a')
````

````python
UTN=[PyQuery(SS[i]).text() for i in range(len(SS))]
````

````python
UT=[URL+PyQuery(SS[i]).attr('href') for i in range(len(SS))]
````
