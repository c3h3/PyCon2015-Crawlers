```python
from pyquery import PyQuery
```

```python
s=PyQuery(res.text)
```

```python
S=s("li.clearfix h2 a")
```

```python
c=[]
page=1
while len(S)!=0:
    url="https://kktix.com/events?page=%s"%page
    res=requests.get(url)
    S=PyQuery(res.text)("li.clearfix h2 a")
    for i in range(len(S)):
        c.append(S[i].text)
    page+=1    
```

```python
import pandas as pd
```

```python
dfs=pd.DataFrame(c)
```

```python
dfs.columns=[['events']]
dfs
```
